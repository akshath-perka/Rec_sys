# create a recommender based on entered genre

```python
movies_df['genres'].value_counts()
```

```python
import ast

def simplify_genres(genres_str):
    genres_list = ast.literal_eval(genres_str)
    genre_names = [genre['name'] for genre in genres_list]
    return ', '.join(genre_names)

# Apply the simplify_genres function to the Genres column
movies_df['genres'] = movies_df['genres'].apply(simplify_genres)
```

```python
movies_df.head()
```

```python
movies_df['cast'].value_counts()
```

```python
def simplify_cast(cast_str):
    cast_list = ast.literal_eval(cast_str)
    cast_names = [cast['name'] for cast in cast_list]
    return ', '.join(cast_names)

# Apply the simplify_genres function to the Genres column
movies_df['cast'] = movies_df['cast'].apply(simplify_cast)
```

```python
def extract_director_names(crew_str):
    crew_list = ast.literal_eval(crew_str)
    directors = [crew['name'] for crew in crew_list if crew['job'] == 'Director']
    return ', '.join(directors)

# Apply the function to the "crew" column
movies_df['crew'] = movies_df['crew'].apply(extract_director_names)
```

```python
movies_df.rename(columns = {'crew':'director'},inplace = True)
```

```python
def simplify_keys(keys_str):
    keys_list = ast.literal_eval(keys_str)
    keys_names = [keywords['name'] for keywords in keys_list]
    return ', '.join(keys_names)

# Apply the simplify_genres function to the Genres column
movies_df['keywords'] = movies_df['keywords'].apply(simplify_keys)
```

```python
def get_recs(movies_df, preferred_genre, top_n=20):
    # Filter movies based on the preferred genre
    filtered_movies = movies_df[movies_df['genres'].str.contains(preferred_genre, case=False, na=False)]
    
    # Sort filtered movies by score in descending order
    sorted_movies = filtered_movies.sort_values(by='score', ascending=False)
    
    # Get the top N recommendations
    top_n_recommendations = sorted_movies.head(top_n)
    
    return top_n_recommendations[['original_title','score','genres']]
```

```python
user_genre = input("Enter your preferred genre: ")

# Get recommendations for the user's preferred genre
recommendations = get_recs(movies_df, user_genre)
recommendations_df = pd.DataFrame(recommendations)
# Display the recommended movies
print("Top 20 Movies in the", user_genre, "genre:")
recommendations_df
```

```python
def get_recs_mult(df, preferred_genres, top_n=20):
    # Split the preferred genres into a list
    preferred_genres = preferred_genres.split(', ')
    
    # Initialize a mask to filter movies
    mask = None
    
    # Create a mask for each preferred genre and combine them
    for genre in preferred_genres:
        genre_mask = df['genres'].str.contains(genre, case=False, na=False)
        if mask is None:
            mask = genre_mask
        else:
            mask &= genre_mask
    
    # Filter movies based on the combined mask
    filtered_movies = df[mask]
    
    # Sort filtered movies by score in descending order
    sorted_movies = filtered_movies.sort_values(by='score', ascending=False)
    
    # Get the top N recommendations
    top_n_recommendations = sorted_movies.head(top_n)
    
    # Return the entire DataFrame with all columns
    return top_n_recommendations[['original_title','score','genres']]
```

```python
user_genres = input("Enter your preferred genres (comma-separated): ")

# Get recommendations for the user's preferred genres
recommendations = get_recs_mult(movies_df, user_genres)
recommendations_df = pd.DataFrame(recommendations)
# Display the recommended movies
print("Top 20 Movies in the genres:", user_genres)
recommendations_df
```

# now based on keywords

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the keywords column to get TF-IDF features
tfidf_matrix = tfidf_vectorizer.fit_transform(movies_df['keywords'])

# Calculate cosine similarity between movies
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Function to get movie recommendations based on input movie title
def get_recs_keys(input_title, cosine_sim=cosine_sim, df=movies_df, top_n=10):
    # Get the index of the movie with the given title
    input_title = input_title.lower()
    
    # Convert all titles in the DataFrame to lowercase
    df['original_title'] = df['original_title'].apply(lambda x: x.lower())
    idx = df.index[df['original_title'] == input_title].tolist()[0]
    
    # Get the pairwise cosine similarity scores with other movies
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort the movies based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the top N similar movies (excluding the input movie)
    sim_scores = sim_scores[1:(top_n+1)]
    
    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]
    
    # Return the top N recommended movies
    return df['original_title'].iloc[movie_indices]
```

```python
user_movie = input("Enter your preferred movie: ")

# Get recommendations for the user's preferred genres
recommendations = get_recs_keys(user_movie)
recommendations_df = pd.DataFrame(recommendations)
# Display the recommended movies
print("Top 20 Movies similar to:", user_movie)
recommendations_df
```
